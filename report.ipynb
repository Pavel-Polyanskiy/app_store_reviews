```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_with_sentiments.csv')
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
plt.figure(figsize = (10, 8))
df.groupby('app_name')['rating'].mean().sort_values().plot(kind = 'barh')
plt.xlabel('Rating')
plt.ylabel('App')
plt.title('Average rating of Crypto-related Apps in App Store')
```

```python
plt.figure(figsize = (10, 8))
df.groupby('app_name')['sentiment_score_vader'].mean().sort_values().plot(kind = 'barh')
plt.xlabel('Sentiment score')
plt.ylabel('App')
plt.title('Average vader sentiment score of Crypto-related Apps in App Store')
```

```python
plt.figure(figsize = (10, 8))
df.groupby('app_name')['sentiment_score_textblob'].mean().sort_values().plot(kind = 'barh')
plt.xlabel('Sentiment score')
plt.ylabel('App')
plt.title('Average TextBlob sentiment score of Crypto-related Apps in App Store')
```
